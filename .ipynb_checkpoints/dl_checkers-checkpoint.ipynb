```python
class Actor():
    def play_move(self, game):
        pass
    
class RandomActor(Actor):
    def play_move(self, game):
        moves = game.get_possible_moves()
        move = randint(0,len(moves)-1)
        game.move(moves[move])
        
class MinMaxActor(Actor):
    def play_move(self, game):
        temp = minMax2(game)
        b = temp[0]
        game.board = b.board
```

```python
def play_game(p1, p2, number_of_games):
    p1_wins = []
    p2_wins = []
    no_wins = []
    
    
    for i in range(number_of_games):
        game = Game()
        b = game.board
        b.maxDepth = 5
        
        while not game.is_over():
            if game.whose_turn() == 1:
                print(game.get_possible_moves())
                p1.play_move(game)
                print_board(game.board)
            else:
                print(game.get_possible_moves())
                p2.play_move(game)
                print_board(game.board)
        if game.get_winner() == 1:
            p1_wins.append(len(game.moves))
        elif game.get_winner() == 2:
            p2_wins.append(len(game.moves))
        else:
            no_wins.append(len(game.moves))
        print("Player {} has won the game after {} moves".format(game.get_winner(), len(game.moves)))
        
    print("Player 1: Wins {} | Average moves {}".format(len(p1_wins), mean(p1_wins)))
    print("Player 2: Wins {} | Average moves {}".format(len(p2_wins), mean(p2_wins)))
    print("No winner: {} times | Average moves {}".format(len(no_wins), mean(no_wins)))
```

```python
p1 = MinMaxActor()
p2 = MinMaxActor()

play_game(p1,p2,100)
```
